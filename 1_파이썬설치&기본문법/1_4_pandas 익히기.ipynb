```python
# 예제 1-32 필터링하기 2
sample_1[condition]
```

```python
# 예제 1-34 데이터 확인하기 
sample_1
```

```python
# 예제 1-35 두 개 컬럼에 필터링하기 1
conditions = (sample_1['성별'] == '남성') & (sample_1['입국객수'] >= 150000)
conditions
```

```python
# 예제 1-36 두 개 컬럼에 필터링하기 2
sample_1[conditions]
```

```python
# 예제 1-37 두 개 컬럼에 필터링하기 3 - 가독성 좋게 표현하기 
conditions = (sample_1['성별'] == '남성') \
    & (sample_1['입국객수'] >= 150000)

sample_1[conditions]
```

```python
# 예제 1-38 한 컬럼에 여러 조건 필터링하기 1
conditions = (sample_1['국적코드'] == 'A01') \
    | (sample_1['국적코드'] == 'A18')
sample_1[conditions]
```

```python
# 예제 1-39 한 컬럼에 여러 조건 필터링하기 2 
conditions = (sample_1['국적코드'].isin(['A01', 'A18'])) 
conditions 
```

```python
# 예제 1-40 한 컬럼에 여러 조건 필터링하기 3
sample_1[conditions]
```

```python
# 예제 1-41 한 컬럼에 여러 조건 필터링하기 4
conditions = (sample_1['국적코드'].isin(['A01', 'A18'])) 
sample_1[conditions == False]
```

## 1.4.5. 데이터 통합하기① - 옆으로 통합(pd.merge)

```python
# 예제 1-42 데이터 확인하기 
sample_1
```

```python
# 예제 1-43 국적코드표 엑셀파일 불러오기 
code_master = pd.read_excel('./files/sample_codemaster.xlsx')
code_master
```

```python
# 예제 1-44 데이터 옆으로 통합하기(left 조건) 
sample_1_code = pd.merge(left=sample_1, 
                         right=code_master,
                         how='left',
                         left_on='국적코드',
                         right_on='국적코드')
sample_1_code
```

```python
# 예제 1-45 데이터 옆으로 통합하기(inner 조건) 
sample_1_code_inner = pd.merge(left=sample_1, 
                               right=code_master,
                               how='inner',
                               left_on='국적코드',
                               right_on='국적코드')
sample_1_code_inner
```

## 1.4.6. 데이터 통합하기② - 아래로 통합(pd.append)

```python
# 예제 1-46 sample_2_code 만들기
sample_2 = pd.read_excel('./files/sample_2.xlsx', 
                         header=1, 
                         skipfooter=2, 
                         usecols='A:C')
sample_2['기준년월']='2018-12'
sample_2_code = pd.merge(left=sample_2, 
                         right=code_master,
                         how='left',
                         left_on='국적코드',
                         right_on='국적코드')
sample_2_code
```

```python
# 예제 1-47 sample_1_code 확인하기 
sample_1_code
```

```python
# 예제 1-48 데이터 아래로 통합하기 1
sample = sample_1_code.append(sample_2_code, ignore_index=True)
sample
```

```python
# 예제 1-49 데이터 아래로 통합하기 2
sample_1_code.append(sample_2_code)
```

## 1.4.8. 데이터 집계하기(pivot_table)

```python
# 예제 1-53 pivot_table() 함수 활용하기 1
sample_pivot = sample.pivot_table(values='입국객수'
                                , index='국적명'
                                , columns='기준년월'
                                , aggfunc='mean'
                                )
sample_pivot
```

```python
# 예제 1-54 pivot_table() 함수 활용하기 2
sample_pivot_2 = sample.pivot_table(values='입국객수'
                                   , index='국적명'
                                   , aggfunc='max')
sample_pivot_2
```
